Real-world Application: **Whether to fish Ilish this year.**

In Bangladesh, Ilish is a very popular fish and is the national fish of the country. Every year, in the rainy season the fishermen catch this fishes. Due to the demand and popularity of this fish, overfishing is rampant. We need to decide what proportion of ilishes to catch in a year so that it maximizes the longer term return. Each ilish generates a fixed amount of dollar. But if a large proportion of ilishes are caught in a specific year, it lowers the fishes for the next year. For this, we have to find the optimum portion of ilishes to catch to maximize the return over a long time period. We can esily formulate this problem as a Markov Decision Process (MDP) like below.

**States:** The number of ilishes available in Bangladesh in a year. For simplicity, we can assume there are only four states: empty, low, medium and high. The four states are defined as follows. **Empty** means no ilishes are available, **low** means available number of ilishes are below a certain threshold n1, **medium** means available number of salmons are between n1 and n2 and **high** means available number of salmons are more than n2.

**Actions:** For simplicity we can assume there are only two actions: 'fish' and 'No fish'. Fish means catching certain proportions of ilishes and 'No fish' means catching no fish. For the **empty** state, the only possible action is 'No fish'. 

**Rewards:** Fishing ilishes at a certain state produces some rewards. We can  assume the rewards of fishing at state low, medium and high are 10K, 100K and 500k respectively. If a 'fish' action takes at empty state then the reward is negative -1000K because it requires re-breeding new ilishes that takes time and money.

**State Transitions:** Below figure is the state transition graph for this MDP problem (Please purdon my hand-drawn figure, was a bit sort of time to draw in computer). Here, there are two types of nodes: the bigger circles are for the four states and the black dots are the actions. Fishing in a state has a higher probability to move to a state with lower number of ilishes. Similarly, no fish action has higher probability to move to a state with higher number of ilishes. After taking an action, the environment responds with a reward and moves to the next state. These state transitions are represented by arrows from the action node to the state nodes. Each arrow shows the transition probability and reward. For example, from the state **High**, 'Fish' action node has 2 arrows transitioning to 2 different states. One is the **Medium** state with probability 0.7 and reward 500K and the other is back to the **High** itself with probability 0.25 and reward 500K. Similarly, it has two transitions for the 'No fish' action. One is back to the **High** itself with probability 0.95 and reward 0k and the other is the **Medium** state with probability 0.05 and reward 0K (in case some fishes die for the overpopulation). In the **Empty** state , only 'Re-breed' action can be taken that transitions to the state **Low** with probability 1 and reward -1000K.
